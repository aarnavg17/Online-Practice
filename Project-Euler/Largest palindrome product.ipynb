```
import math
largest_palindrome=int()
product=int()
reverse=int()
for i in range(100,1000):
  for j in range (100, 1000):
    product=i*j
    p=product
    reverse=0
    while product!=0:
      reverse=(reverse*10) + (product%10)
      product=math.floor(product/10)
    if reverse == p:
      if reverse>largest_palindrome:
        largest_palindrome=reverse
print (largest_palindrome)
```
