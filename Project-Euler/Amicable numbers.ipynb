```
import math
import time
start=time.time()

def d (number):
  add=1
  for i in range (2,math.floor(number**0.5)+1):
    if number%i==0:
      add+=(i+(number/i))
  return add

sum_amicable=0
num_list=[]
for num in range (1,10000):
  if num in num_list:
    continue
  sum1=d(num)
  if sum1 <10000 and sum1!=num:
    sum2=d(sum1)
    if sum2==num:
      sum_amicable+=(sum1+num)
      num_list.append(sum1)

print(sum_amicable)
print(time.time()-start)
```
