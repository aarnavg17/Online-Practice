```
from google.colab import drive
drive.mount('/content/drive')
import time
start=time.time()
with open ('/content/drive/My Drive/Python/Project Euler/p022_names.txt') as f:
  R=f.read()
name_list=R.split(",")
name_list.sort()
sum1,sum2=0,0
num=1
for i in name_list:
  for j in range(1,len(i)-1):
    sum1+=(ord(i[j])-64)
  sum2+=(sum1*num)
  num+=1
  sum1=0
print(sum2)
print(time.time()-start)
```
