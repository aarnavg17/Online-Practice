```
import math
number=600851475143
largest_prime_factor=int()
for n in range (3, math.ceil(math.sqrt(number)),2):
  if number%n==0:
    not_prime=0
    for i in range (3,math.ceil(math.sqrt(number)),2):
      if n%i==0:
        not_prime=1
    if not_prime!=1:
      largest_prime_factor=n
print(largest_prime_factor)
```
