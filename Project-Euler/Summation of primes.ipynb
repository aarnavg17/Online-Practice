```
import time
start=time.time()
#import math
sum_prime=17
i=9
flag=False
while i<2000000:
  i+=2
  flag=True
  for j in range (3, int(i**0.5)+1,2):
    if i%j==0:
      flag=False
      break
  if flag:
    sum_prime+=i
print(sum_prime)
print(time.time()-start)
```

```
import time
start=time.time()
limit = 2000000
numList = []
sumOfPrimes = 0

for element in range(2, limit):
    numList.append(element)
for item in range(2, limit):
    if numList[item - 2] != -1:
        sumOfPrimes += numList[item - 2]
        c = 2 * item
        while c < limit:
            numList[c - 2] = -1
            c += item

print(sumOfPrimes)
print(time.time()-start)
```
