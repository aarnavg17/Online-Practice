```
import math
import time
start=time.time()
finalcount, finalnum=1,1
num_list = [False,True]*500
num_list[1]=False
result=[]
for num in range (3,1000,2):
    for n in range (num**2,1000,num):
        num_list[n]=False
for x in range (3,len(num_list)):
    if num_list[x]:
        result.append(x)
for num in result:
    testrun,a=0,1
    temp_list=[x for x in range(0,num)]
    while True:
        a=(a*10)%num
        testrun+=1
        if temp_list[a]!=a:
            if testrun+temp_list[a]>finalcount:
                finalcount=testrun+temp_list[a]
                finalnum=num
            break
        else:
            temp_list[a]=testrun*(-1)
print(finalnum)
print(time.time()-start)
```
