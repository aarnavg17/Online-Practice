```
#HackerRank
pastlist=[0]*5000001
list2=[0]*5000001
def longcol(N):
    listn=[]
    long_chain=1
    for n in range (3,N,2):
        chain=1
        num=n
        while n!=1:
            listn.append(n)
            if pastlist[n]:
                chain+=pastlist[n]-1
                break
            list2[n]=chain
            if n%2:
                n=(3*n)+1
            else:
                n//=2
            chain+=1
        for i in listn:
            pastlist[i]=chain-list2[i]
        pastlist[num]=chain
        if chain > long_chain:
            final_num=num
            long_chain=chain
    print(final_num)

T=int(input())
for i in range (0,T):
    N=int(input())
    longcol(N)
```

```
#HackerRank attempt 2

past = {9 : 19, 13 : 10}

def longcol2(N):
    long_chain = 1
    for i in range (3, N , 2):
        if i in past.keys():
            chain = past[i]
        else:
            chain = 1
            num = i
            while i!=1:
                if i in past.keys():
                    chain += past[i]
                else:
                    if i % 2:
                        i = (3 * i) + 1
                    else:
                        i /= 2
                    chain += 1
            past.update({num : chain})
        if chain >= long_chain:
            final_num = num
            long_chain = chain
    return final_num

T = int(input())
for i in range (0, T):
    N = int(input())
    print(longcol2(N))
```
