```
import time
start=time.time()
first_sunday=0
date=366%7
for year in range (1901,2001):
  if (year/4==0 and year/100!=0) or year/400==0:
    for month in (31,29,31,30,31,30,31,31,30,31,30,31):
      date+=month
      date%=7
      if date==7:
        first_sunday+=1
  else:
    for month in (31,28,31,30,31,30,31,31,30,31,30,31):
      date+=month
      date%=7
      if date/7==0:
        first_sunday+=1
print(first_sunday)
print(time.time()-start)
```
