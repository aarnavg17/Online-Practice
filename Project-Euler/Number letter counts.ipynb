```
import time
start=time.time()

def digit(d):
  if d==1 or d==2 or d==6:
    return 3
  elif d==4 or d==5 or d==9:
    return 4
  elif d==0:
    return 0
  else:
    return 5

def dig2(d): #ty
  if d==4 or d==5 or d==6:
    return 3
  elif d==2 or d==3 or d==8 or d==9:
    return 4
  else:
    return 5

def dig3(d): #teen
  if d==5 or d==6:
    return 3
  elif d==3 or d==4 or d==8 or d==9:
    return 4
  else:
    return 5

letters=11 #"one thousand"
for i in range (1,1000):
  print(i)
  if i >100:
    letters+=7 #"hundred"
    letters+=digit(int(str(i)[:1])) #kitna hundred
    if i%100:
      letters+=3 #"and"
    i%=100
  elif i==100:
    letters+=10
    continue
  if i>=20:
    letters+=2 #"ty"
    letters+=digit(i%10)
    letters+=dig2(int(str(i)[:1]))
  elif i>=13:
    letters+=4 #"teen"
    letters+=dig3(i%10)
  elif i==11 or i==12:
    letters+=6
  elif i==10:
    letters+=3
  else:
    letters+=digit(i)
  print(letters)
print(letters)
print(time.time()-start)
```
