```
#Runtime error
nums = [2, 5, 1, 3, 4, 7]
n = 3
shuffle = []
shot, i = 0, 0
for _ in range(len(nums)):
    print ("nums: ", nums)
    print ("shuffle: ", shuffle)
    print ("i: ", i)
    print("shot: ", shot)
    shuffle.append(nums[i])
    i += (((-1) ** shot) * n) + (1 * shot)
    shot = int(not shot)
print (shuffle)
```

```
#Fucking Pythonify
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        res= []
        for i in range(n):
            res+= nums[i], nums[i+n] 
        return res
```
