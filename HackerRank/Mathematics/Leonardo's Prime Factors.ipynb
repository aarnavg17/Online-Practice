```
def number_of_primes(n):
    prime_list = [2, 3]
    x = 3
    while len(prime_list) != n:
        x += 2
        flag = True
        for i in range(3, int(x ** 0.5) + 1, 2):
            if not (x % i):
                flag = False
                break
        if flag:
            prime_list.append(x)
    return(prime_list)

for _ in range(int(input())):
    n = int(input())
    if n == 1:
        print("0")
    elif n <= 5:
        print("1")
    else:
        prime_list = number_of_primes(16)
        count, prod = 0, 1
        for x in prime_list:
            count += 1
            prod *= x
            if prod > n:
                print(count - 1)
                break
```
