```
M = int(input())
M_list = input().split()
N = int(input())
N_list = input().split()

M_set = set(M_list)
N_set = set(N_list)

out = M_set.symmetric_difference(N_set)
#print(out)
out_list = list(out)
#print(out_list)
for i in range(len(out_list)):
    out_list[i] = int(out_list[i])
out_list.sort()
#print(out_list)
for x in out_list:
    print(x)
```
