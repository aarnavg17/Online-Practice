```
def check(UID):
    uppercase = 0
    digit = 0
    flag = True
    if UID.isalnum() and len(UID) == 10:
        for x in UID:
            if UID.count(x) > 1:
                return "Invalid"
            if x.isdigit():
                digit += 1
            elif x.isupper():
                uppercase += 1
    else:
        return "Invalid"
    if digit >= 3 and uppercase >= 2:
            return "Valid"
    else:
        return "Invalid"


for _ in range (int(input())):
    UID = input()
    print(check(UID))
```
