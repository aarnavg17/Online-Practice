```
records_dict = {}

for _ in range(int(input())):
    name = input()
    score = float(input())
    if score in records_dict.keys():
        records_dict[score].append(name)
    else:
        records_dict[score] = [name]

marks = list(set(records_dict.keys()))
marks.sort()
second_lowest = marks[1]
records_dict[second_lowest].sort()
for x in records_dict[second_lowest]:
    print(x)
```
