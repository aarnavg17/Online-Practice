```
def check(CC):
    if CC[0] != "4" and CC[0] != "5" and CC[0] != "6":
        return "Invalid"
    else:
        for i in range(13):
            if CC[i] == CC[i + 1] == CC[i + 2] == CC[i + 3]:
                return "Invalid"
    return "Valid"

def valid(CC):
    if len(CC) != 16 and len(CC) != 19:
        return "Invalid"
    elif len(CC) == 19:
        if CC[4] != "-" or CC[9] != "-" or CC[14] != "-" or not CC[0:4].isdigit() or not CC[5:9].isdigit() or not CC[10:14].isdigit() or not CC[15:].isdigit():
            return "Invalid"
        else:
            CC2 = ""
            for x in CC:
                if x != "-":
                    CC2 += x
            return check(CC2)
    elif CC.isdigit():
        return check(CC)

for _ in range (int(input())):
    CC = input()
    print(valid(CC))
```
